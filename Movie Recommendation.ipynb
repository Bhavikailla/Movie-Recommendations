```python
import matplotlib.pyplot as plt
data.id.value_counts(dropna=False).sort_index().plot(kind='bar', figsize=(15,16))
plt.show()
```

```python
data = data.loc[:,['id','title','year','rating']]
```

```python
data['id'] = data['id'].replace(np.nan, 'unknown')
data['title'] = data['title'].replace(np.nan, 'unknown')
data['year'] = data['year'].replace(np.nan, 'unknown')
data['rating'] = data['rating'].replace(np.nan, 'unknown')
```

```python
data
```

```python
data['title'] = data['title'].str.replace('|', ' ')
```

```python
data
```

```python
data['title'] = data['title'].str.lower()
```

```python
# null terminating char at the end
data['title'][1718]
```

```python
# removing the null terminating char at the end
data['title'] = data['title'].apply(lambda x : x[:-1])
```

```python
data.to_csv('data.csv',index=False)
```
